```python
## TRAIT GENERATION

TOTAL_BANANAS = 100

traits = []

def createCombo():
    
    trait = {}

    trait["Background"] = random.choices(backgrounds, backgroundweights)[0]
    trait["Pumpkin"] = random.choices(bananas, bananaweights)[0]
    trait["Stone"] = random.choices(mouths, mouthweights)[0]
    trait["Animal"] = random.choices(animals, animalweights)[0]
    trait["Attribute"] = random.choices(eyes, eyesweights)[0]
    trait["Hat"] = random.choices(headgears, headgearweights)[0]
    
    if trait in traits:
#         if this Banana was already in the Traits list, restart the function
        return createCombo()
    else:
        return trait
    
for i in range(TOTAL_BANANAS):
    
    newtraitcombo = createCombo()
#     newtraitcombo
    
    traits.append(newtraitcombo)
    
```

```python
## ARE ALL BANANAS UNIQUE? I DUNNO HOW THIS WORKS BUT IT WORKS
def allUnique(x):
    seen = list()
    return not any(i in seen or seen.append(i) for i in x)

print(allUnique(traits))
```

```python
# ADD TOKEN IDS TO JSON

i = 0
for item in traits:
    item["tokenId"] = i
    i = i + 1
```

```python
# GET TRAIT COUNTS

backgroundcounts = {}
for item in backgrounds:
    backgroundcounts[item] = 0
    
bananacounts = {}
for item in bananas:
    bananacounts[item] = 0
    
mouthcounts = {}
for item in mouths:
    mouthcounts[item] = 0
    
eyescounts = {}
for item in eyes:
    eyescounts[item] = 0

animalcounts = {}
for item in animals:
    animalcounts[item] = 0
    
    
headgearcounts = {}
for item in headgears:
    headgearcounts[item] = 0
    
for banana in traits:
#     print(banana)
    backgroundcounts[banana["Background"]] += 1
    bananacounts[banana["Pumpkin"]] += 1
    mouthcounts[banana["Stone"]] += 1
    animalcounts[banana["Animal"]] += 1
    eyescounts[banana["Attribute"]] += 1
    headgearcounts[banana["Hat"]] += 1
    

    
print(backgroundcounts)
print(bananacounts)
print(mouthcounts)
print(eyescounts)
print(headgearcounts)
```

```python
  
    
#### IMAGE GENERATION
    
for item in traits:

    im1 = Image.open(f'./Background/{backgroundfiles[item["Background"]]}.png').convert('RGBA')
    im2 = Image.open(f'./pumpkins/{bananafiles[item["Pumpkin"]]}.png').convert('RGBA')
    im3 = Image.open(f'./stone/{mouthfiles[item["Stone"]]}.png').convert('RGBA')
    im6 = Image.open(f'./animals/{animalfiles[item["Animal"]]}.png').convert('RGBA')
    im4 = Image.open(f'./attribute/{eyefiles[item["Attribute"]]}.png').convert('RGBA')
    im5 = Image.open(f'./hats/{headgearfiles[item["Hat"]]}.png').convert('RGBA')


    #Create each composite
    com1 = Image.alpha_composite(im1, im3)
    com2 = Image.alpha_composite(com1, im2)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)
    com5 = Image.alpha_composite(com4, im6)


    #Convert to RGB
    rgb_im = com5.convert('RGB')
    #display(rgb_im.resize((400,400), Image.NEAREST))
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./output/" + file_name)
    
    

    
```
